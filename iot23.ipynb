```python
import pandas as pd
import os
pd.options.display.max_rows = 300
pd.options.display.max_columns = 300
```

```python
basepath = 'D:/iot/IoTScenarios'
path = []
df=[]
for i in os.listdir('/iot/IoTScenarios/'):
    path = f"{basepath}/{i}"
    path = f'{path}/{os.listdir(path)[0]}'
#     print(path)
    if 'conn.log.labeled' in os.listdir(path)[0]:
        path = f"{path}/{os.listdir(path)[0]}"
        a = pd.read_table(filepath_or_buffer=path, skiprows=10, nrows=100000)
        a.columns=['ts',
              'uid',
              'id.orig_h',
              'id.orig_p',
              'id.resp_h',
              'id.resp_p',
              'proto',
              'service',
              'duration',
              'orig_bytes',
              'resp_bytes',
              'conn_state',
              'local_orig',
              'local_resp',
              'missed_bytes',
              'history',
              'orig_pkts',
              'orig_ip_bytes',
              'resp_pkts',
              'resp_ip_bytes',
              'label']
        a.drop(a.tail(1).index,inplace=True)
        df.append(a)
        del a
    else:
        path = f"{path}/{os.listdir(path)[0]}"
        path = f"{path}/{os.listdir(path)[0]}"
        a = pd.read_table(filepath_or_buffer=path, skiprows=10, nrows=100000)
        a.columns=['ts',
              'uid',
              'id.orig_h',
              'id.orig_p',
              'id.resp_h',
              'id.resp_p',
              'proto',
              'service',
              'duration',
              'orig_bytes',
              'resp_bytes',
              'conn_state',
              'local_orig',
              'local_resp',
              'missed_bytes',
              'history',
              'orig_pkts',
              'orig_ip_bytes',
              'resp_pkts',
              'resp_ip_bytes',
              'label']
        a.drop(a.tail(1).index,inplace=True)
        df.append(a)
        del a
```

```python
data = pd.concat(df, axis=0)
```
